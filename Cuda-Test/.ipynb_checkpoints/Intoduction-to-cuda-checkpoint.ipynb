This notebook is going to show you beginner friendly examples of CUDA syntax. 

Starting with declarating a __global__ kernel that runs on the GPU and adds two numbers

```c++
__global__ void add( int a, int b, int *c ) {
 *c = a + b;
}
```

Next we have to run this kernel. The `<<< >>>` syntax helps us with that. The first number is how many blocks and the second is how many threads we are going to use.

Before everything we need to allocate vram with cudaMalloc() on the gpu to store the result of both added numbers. Then we run the kernel with 2 and 7.

Next we need to copy that vram memmory to the host. We do that with cudaMemcpy(). And finally we print the final number and free the memmory

```c++
#include <iostream>

int c;
int *d_c;

cudaMalloc( (void**)&d_c, sizeof(int) );

add<<<1,1>>>( 2, 7, d_c );

 cudaMemcpy( &c, d_c, sizeof(int), cudaMemcpyDeviceToHost );

 printf( "2 + 7 = %d\n", c );
 cudaFree( d_c );
 
```

Next we are going to try something a little bit more difficult. We are going to add two vectors index by index and store the result in a third vector. Firstly you can see a CPU implementation of this, where a function adds the numbers one by one which is slow.

```c++
#include <chrono>

const int N = 20;

void add_CPU( int *a, int *b, int *c ) {
     int tid = 0; // this is CPU zero, so we start at zero
     while (tid < N) {
         c[tid] = a[tid] + b[tid];
         tid += 1; // we have one CPU, so we increment by one
     }
}

int cpu_a[N], cpu_b[N], cpu_c[N];

// fill the arrays 'a' and 'b' on the CPU
for (int i=0; i<N; i++) {
    cpu_a[i] = i;
    cpu_b[i] = i * i;
}

auto start = std::chrono::high_resolution_clock::now();

add_CPU( cpu_a, cpu_b, cpu_c );

auto end = std::chrono::high_resolution_clock::now();

std::chrono::duration<double, std::milli> CPU_time = end - start;

// display the results
for (int i=0; i<N; i++) {
    printf( "%d + %d = %d\n", cpu_a[i], cpu_b[i], cpu_c[i] );
}
```

Next you will see how to add two vectors faster using parallel programming.

Before any code you can see a visualization of the addition.

```c++
#include "assets/display.hpp"

show_image("assets/vectoradd.png", 25);
```

First we need the global function. As you can see the kernel prefforms only one addition. In order to add the whole vector we need multiple blocks running at the same time.

```c++
__global__ void vector_add( int *a, int *b, int *c ) {
     int tid = blockIdx.x; // handle the data at this index
     if (tid < N)
     c[tid] = a[tid] + b[tid];
}
```

Parallel programming is what makes the GPU great for a big number of the same operations being performed simultaneously on different pieces of data.

We can see that we call an **N** number of krernels to preform only one addition in the vector each. In this example we don`t save that much time but parallel programming works in our favour because when we go to in to the millions of operations we save a lot of time instead of adding them one by one with the CPU.

Bellow you can see how blocks and threads are spaced out in the grid.

```c++
show_image("assets/CudaGrid.png", 50);
```

```c++
int v_a[N], v_b[N], v_c[N];
int *dev_a, *dev_b, *dev_c;

// allocate the memory on the GPU
cudaMalloc( (void**)&dev_a, N * sizeof(int) );
cudaMalloc( (void**)&dev_b, N * sizeof(int) );
cudaMalloc( (void**)&dev_c, N * sizeof(int) );

// fill the arrays 'a' and 'b' on the CPU
for (int i=0; i<N; i++) {
    v_a[i] = i;
    v_b[i] = i * i;
}

// copy the arrays 'a' and 'b' to the GPU
cudaMemcpy( dev_a, v_a, N * sizeof(int), cudaMemcpyHostToDevice );
cudaMemcpy( dev_b, v_b, N * sizeof(int), cudaMemcpyHostToDevice );


float GPU_time = 0;
cudaEvent_t start, stop;

cudaEventCreate(&start);
cudaEventCreate(&stop);

// Record start event on the default stream
cudaEventRecord(start);

vector_add<<<N,1>>>( dev_a, dev_b, dev_c );

cudaEventRecord(stop);

// CRITICAL: Wait for the GPU to finish before calculating
cudaEventSynchronize(stop);

cudaEventElapsedTime(&GPU_time, start, stop);

cudaEventDestroy(start);
cudaEventDestroy(stop);

// copy the array 'c' back from the GPU to the CPU
cudaMemcpy( v_c, dev_c, N * sizeof(int), cudaMemcpyDeviceToHost );

// display the results
for (int i=0; i<N; i++) {
    printf( "%d + %d = %d\n", v_a[i], v_b[i], v_c[i] );
}

// free the memory allocated on the GPU
cudaFree( dev_a );
cudaFree( dev_b );
cudaFree( dev_c );
```

```c++
printf("Time CPU: ", CPU_time);
printf("Time GPU: ", GPU_time);
```

# Generating a Fractal (The Julia Set)

We are using CUDA to draw a complex mathematical shape called a Julia Set. It applies the same formula and colors each pixel based on the value. The formula used for this example is:$$Z_{n+1} = Z_n^2 + C$$ Instead of calculating pixels one by one on the CPU, we launch a kernel where each GPU thread calculates the color of a single pixel simultaneously. How it works:cuComplex: A helper to do math with complex numbers ($a + bi$).julia: The math logic. It checks if a point "escapes" to infinity or stays trapped in the set.julia_kernel: The GPU function that tells each pixel: "If you're in the set, turn Red ($255$); if not, stay Black ($0$)."

```c++
#include <cuda_runtime.h>

#define DIM 1000

struct cuComplex {
    float r;
    float i;
    __host__ __device__ cuComplex(float a, float b) : r(a), i(b) {}
    __device__ float magnitude2(void) { return r * r + i * i; }
    __device__ cuComplex operator*(const cuComplex& a) {
        return cuComplex(r*a.r - i*a.i, i*a.r + r*a.i);
    }
    __device__ cuComplex operator+(const cuComplex& a) {
        return cuComplex(r+a.r, i+a.i);
    }
};

__device__ int julia(int x, int y) {
    const float scale = 1.5;
    float jx = scale * (float)(DIM/2 - x)/(DIM/2);
    float jy = scale * (float)(DIM/2 - y)/(DIM/2);
    cuComplex c(-0.8, 0.156);
    cuComplex a(jx, jy);
    for (int i=0; i<200; i++) {
        a = a * a + c;
        if (a.magnitude2() > 1000) return 0;
    }
    return 1;
}

__global__ void julia_kernel(unsigned char *ptr) {
    int x = blockIdx.x;
    int y = blockIdx.y;
    int offset = x + y * gridDim.x;
    int juliaValue = julia(x, y);
    ptr[offset*4 + 0] = 255 * juliaValue;
    ptr[offset*4 + 1] = 0;
    ptr[offset*4 + 2] = 0;
    ptr[offset*4 + 3] = 255;
}
```

Here we call the kernel and write the data to a .png file

```c++
#define STB_IMAGE_WRITE_IMPLEMENTATION
#include "assets/stb_image_write.h"

unsigned char *bitmap;
int size = DIM * DIM * 4;
cudaMallocManaged(&bitmap, size);

dim3 grid(DIM, DIM);
julia_kernel<<<grid, 1>>>(bitmap);

cudaDeviceSynchronize();

stbi_write_png("assets/julia.png", DIM, DIM, 4, bitmap, DIM * 4);
std::cout << "Saved julia.png" << std::endl;

cudaFree(bitmap);
```
