```c++
#undef __noinline__
#include <iostream>
#include <math.h>

__global__ void kernel(int a, float* b)
{
    *b = log((float)a);
}

float* b;
int a = 16;

cudaMallocManaged(&b, sizeof(float));

kernel<<<1,1>>>(a, b);

cudaDeviceSynchronize();

std::cout << "Result: " << *b << std::endl;

cudaFree(b);
```
